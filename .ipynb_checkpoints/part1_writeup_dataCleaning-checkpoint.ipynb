**Data Source:**
The data set is specifically chosen columns from the 1990s Califorina census. Each row is a data point representing one block. So when looking at something like total bedrooms the assumption is total number of bedrooms in that specific block.

**Source:**

Nugent, C. (n.d.). California Housing Prices. Kaggle. Retrieved July 5, 2023, from https://www.kaggle.com/datasets/camnugent/california-housing-prices?resource=download

**Data Attributes:**

There are 20640 rows and 10 colums, 9 of the 10 colums are emprical numbers(Float64)and 1 column (the last one) is a catagorical object. The table is a normal Tabular data in multi-tabel form with a file size of 1.4MB. 

**The columns include:**

longitude (Float64): A measure of how far west a house is; a higher value is farther west

latitude (Float64): A measure of how far north a house is; a higher value is farther north

housing Median Age (Float64): Median age of a house within a block; a lower number is a newer building

total Rooms (Float64): Total number of rooms within a block

total Bedrooms (Float64): Total number of bedrooms within a block

population (Float64): Total number of people residing within a block

households (Float64): Total number of households, a group of people residing within a home unit, for a block

median Income (Float64): Median income for households within a block of houses (measured in tens of thousands of US Dollars)

median House Value (Float64): Median house value for households within a block (measured in US Dollars)

ocean Proximity (Object): Location of the house w.r.t ocean/sea


**Data Cleaning Explination**

First all of the data that has Nan data is dropped since filling by mean or median may cause a bias shift toward the predfined means/medians. Second duplicate values are searched for and dropped since the double count would cause an incorrect shift of the caclulated mean from the true mean. Third the catagorical column 'ocean proximity' is converted into nums so they are easier to deal with when coding, since everything else is a num.

```python
#Importing the Data
housing = pd.read_csv("housing.csv")
```

```python
#Cleaning data

#Checking for Null data and dropping since filling by mean or median would throw off the data
housing = housing.dropna()
print(housing.isnull().sum())

#Checking for dupplicate data but there isn't any so we don't have to worry about this
print("Duplicates",housing.duplicated().sum())

#We could check for outliner and drop them. However, we don't want to for this anaysis because we 
#want to span the largest home price range possible to fit an all encompassing model

#This is where we type cast the ocean_proximity vairbale for ease of use in floats later

#Turning the object data into num type and creating array
encode = OneHotEncoder(sparse = False)
oceanProxReshape = housing['ocean_proximity'].values.reshape(-1, 1)
oceanProxEncode = encode.fit_transform(oceanProxReshape)

#appending new col onto the end of the housing df 
newOceanProx = pd.DataFrame(oceanProxEncode, columns= encode.get_feature_names_out(['ocean_proximity']))

#resetting Indices
newOceanProx = newOceanProx.reset_index(drop=True)

#adding to df
housing = housing.reset_index(drop=True).join(newOceanProx)

#dropping old col
housing = housing.drop(columns='ocean_proximity')

housing.head()
```

```python
columns_to_include = ['longitude', 'latitude', 'housing_median_age', 'total_rooms', 
                      'total_bedrooms', 'population', 'households', 
                      'median_income', 'median_house_value']

housing[columns_to_include].hist(bins=50, figsize=(20,15))
plt.show()
plt.show()

ocean_proximity_columns = ['ocean_proximity_<1H OCEAN', 'ocean_proximity_INLAND', 'ocean_proximity_ISLAND', 'ocean_proximity_NEAR BAY', 'ocean_proximity_NEAR OCEAN']
ocean_prox_count = {}

for name in ocean_proximity_columns:
    ocean_prox_count[name] = housing[name].sum()
    
ocean_prox = pd.Series(ocean_prox_count)

ocean_prox.plot.bar()
plt.title('Ocean proximity catagory frequencies')
plt.xlabel('Ocean proximity catagory')
plt.ylabel('Frequency')
plt.show()
```
