**Addressing Multicollinearity**

```python
#Splitting the data into test and train so we can test the model with the test later so check validity
X = housing.drop(columns=['median_house_value'])
y = housing['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


#Finding the VIF to determine which variable are revalent and which can be explained by other variables
#Calculates VIF
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

#Applying the VIF function to all factors and droping those with VIF larger than 10
vif = calculate_vif(X_train)
dropped = []
while vif["VIF"].max() > 10:
    #Finds the col with VIF larger than 10
    remove = vif.sort_values("VIF", ascending=False).iloc[0]
    dropped.append(remove['Variable'])
    #Removes those variables from the train and test df
    X_train = X_train.drop(columns=[remove['Variable']])
    X_test = X_test.drop(columns=[remove['Variable']])
    vif = calculate_vif(X_train)

names = vif["Variable"]
print("Dropped Variables:\n")
for variable in dropped:
    print(f"\t{variable}")
print("\n")
print("Current Variables:")
print(vif)

# Creating a correlation matrix
corr_matrix = X_train.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

**Models**

```python
#Training Linear Regression Model
linear_reg = LinearRegression()
linear_reg.fit(X_train,y_train)
y_prediction = linear_reg.predict(X_test)
mse = mean_squared_error(y_test, y_prediction)

#Printing performance of the model
print(f"Model Accuracy:{linear_reg.score(X_test,y_test)}")
print(f"Linear Regression Mean Square Error: ${mse}")
print(f"Linear Regression Mean Error: ${math.sqrt(mse)}")
print("\n")

plt.figure()
plt.scatter(y_test, y_prediction)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values for Linear Regression")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
plt.show()

#Plotting most important features to see what's influencing the model
importance = linear_reg.coef_
for i,v in enumerate(importance):
    print(f"Feature: {names[i]}, Score:{v}")
    
#Plotting the coef campared to variable
plt.figure()
plt.barh(names,importance)
plt.ylabel("Variables")
plt.xlabel("Coefficient")
plt.title("Importance of variable in predicting housing price")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

```python
#Training Random Forest Model
forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
forest_model.fit(X_test,y_test)
y_pred_rf = forest_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)

#Printing performance of the model
print(f"Model Accuracy:{forest_model.score(X_test,y_test)}")
print(f"Random Forest Mean Square Error: ${mse_rf}")
print(f"Random Forest Mean Error: ${math.sqrt(mse_rf)}")
print("\n")


# Visualizing one of the trees in the random forest
plt.figure(figsize=(20, 10))
plot_tree(forest_model.estimators_[0], filled=True, feature_names=names, max_depth=2, fontsize=10)
plt.title("Sample Tree from Random Forest")
plt.show()

#Actual vs predicted output plot
plt.figure()
plt.scatter(y_test, y_pred_rf)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values for Linear Regression")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red') # 45-degree line
plt.show()

#See what variables influencing the model the most
importance = forest_model.feature_importances_
for i,v in enumerate(importance):
    print(f"Feature: {names[i]}, Score:{v}")
    
#plotting imporance
plt.figure()
plt.barh(names,importance)
plt.ylabel("Variables")
plt.xlabel("Importance")
plt.title("Importance of variable in predicting housing price")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

```python
#Train Gradient Boosting Model
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)
mse_gb = mean_squared_error(y_test, y_pred_gb)

#Printing performance of the model
print(f"Model Accuracy:{gb_model.score(X_test,y_test)}")
print(f"Gradient Boosting Mean Square Error: ${mse_gb}")
print(f"Gradient Boosting Mean Error: ${math.sqrt(mse_gb)}")
print("\n")

#Loss function for gradient boosting
plt.figure()
plt.plot(gb_model.train_score_)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Training Loss for Gradient Boosting")
plt.show()

#Actual vs predicted output plot
plt.figure()
plt.scatter(y_test, y_pred_gb)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted Values for Linear Regression")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red') # 45-degree line
plt.show()

#Find the most important features in predicting housing price
importance = gb_model.feature_importances_
for i,v in enumerate(importance):
    print(f"Feature: {names[i]}, Score:{v}")
    
#plotting imporance
plt.figure()
plt.barh(names,importance)
plt.ylabel("Variables")
plt.xlabel("Importance")
plt.title("Importance of variable in predicting housing price")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

### Why Random Forest Model Worked Best
The Random Forest model out performed the Linear Regression and Gradient Boosting models because the random forest is best situated for models like the 1990s California Housing price data set.


#### Model Complexity and Ensemble Learning

Random Forest operates using ensemble learning, which means it aggregates results from multiple decision trees to deliver its predictions. This aggregation often results in a model that is both more accurate and robust, particularly when the individual models (trees) have diverse perspectives (predictions). This diversity is brought about by the bootstrapping of the data and random feature selection in Random Forests.

#### Overfitting

While the individual decision trees within a Random Forest can be prone to overfitting, the ensemble nature of Random Forest works to combat this tendency. The bootstrapping process (sampling with replacement) introduces variability, ensuring that each tree within the forest has its nuances. By averaging predictions from all trees (for regression) or using majority voting (for classification), the model generalizes better to unseen data.

#### Robustness

Random Forests are known for their robustness and are generally less sensitive to outliers and potentially irrelevant features. If the dataset contains outlier data points or noisy features, Random Forests can deliver robust predictions by leveraging the collective power of the ensemble to average out the noise.

#### Handling of Non-linearities

Random Forests, being constructed of multiple decision trees, inherently manage non-linear relationships between features and the target variable with ease. If the relationship between the predictors and the California housing prices is not strictly linear, the Random Forest would adeptly capture these non-linearities, offering a performance edge over the Linear Regression model.
